```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display, clear_output
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import accuracy_score, r2_score, confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder

# --- Load & preprocess dataset ---
data = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
data["TotalCharges"] = pd.to_numeric(data["TotalCharges"], errors="coerce")
data.dropna(inplace=True)
le = LabelEncoder()
data["Churn"] = le.fit_transform(data["Churn"])

# Regression setup
X_reg = data[["tenure","SeniorCitizen"]]
y_reg = data["MonthlyCharges"]
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg,y_reg,test_size=0.2,random_state=42)
reg_model = LinearRegression()
reg_model.fit(X_train_reg,y_train_reg)
reg_pred = reg_model.predict(X_test_reg)

# Classification setup
X_cls = data[["tenure","MonthlyCharges","SeniorCitizen"]]
y_cls = data["Churn"]
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(X_cls,y_cls,test_size=0.2,random_state=42)
cls_model = LogisticRegression(max_iter=1000)
cls_model.fit(X_train_cls,y_train_cls)
cls_pred = cls_model.predict(X_test_cls)

# --- Dark theme setup ---
plt.style.use("dark_background")
sns.set_theme(style="darkgrid")

# --- Widgets ---
tenure_slider = widgets.IntSlider(value=24,min=1,max=72,step=1,description="Tenure (months)")
charges_slider = widgets.FloatSlider(value=75,min=0,max=150,step=5,description="Monthly Charges")
senior_toggle = widgets.ToggleButtons(options=[0,1],description="SeniorCitizen")
mode_dropdown = widgets.Dropdown(options=["Regression (MonthlyCharges)","Classification (Churn)"],value="Classification (Churn)",description="Mode")

predict_button = widgets.Button(description="Predict",button_style="success")
metrics_button = widgets.Button(description="Show Metrics",button_style="info")
reset_button = widgets.Button(description="Reset",button_style="warning")

output = widgets.Output()

# --- Functions ---
def update_dashboard(change=None):
    with output:
        clear_output(wait=True)
        if mode_dropdown.value=="Regression (MonthlyCharges)":
            new_customer = pd.DataFrame([[tenure_slider.value, senior_toggle.value]],columns=["tenure","SeniorCitizen"])
            prediction = reg_model.predict(new_customer)[0]
            print(f"💰 Predicted Monthly Charge: {prediction:.2f}")
            print(f"R2 Score: {r2_score(y_test_reg,reg_pred):.2f}")
            
            # Plot regression scatter
            fig, ax = plt.subplots(figsize=(6,4))
            ax.scatter(X_test_reg["tenure"],y_test_reg,color="cyan",label="Actual")
            ax.scatter(X_test_reg["tenure"],reg_pred,color="magenta",label="Predicted",alpha=0.7)
            ax.set_xlabel("Tenure")
            ax.set_ylabel("Monthly Charges")
            ax.set_title("Regression Fit")
            ax.legend()
            plt.show()
            
        else:
            new_customer = pd.DataFrame([[tenure_slider.value, charges_slider.value, senior_toggle.value]],
                                        columns=["tenure","MonthlyCharges","SeniorCitizen"])
            prediction = cls_model.predict(new_customer)[0]
            result = "⚠️ Customer Will Churn" if prediction==1 else "✅ Customer Will Stay"
            print(result)
            print(f"Accuracy: {accuracy_score(y_test_cls,cls_pred):.2f}")
            
            # Confusion matrix
            cm = confusion_matrix(y_test_cls,cls_pred)
            fig, ax = plt.subplots(1,2,figsize=(12,4))
            sns.heatmap(cm,annot=True,fmt="d",cmap="magma",
                        xticklabels=["Stay","Churn"],yticklabels=["Stay","Churn"],ax=ax[0])
            ax[0].set_title("Confusion Matrix")
            
            # Pie chart of churn distribution
            churn_counts = y_cls.value_counts()
            ax[1].pie(churn_counts,labels=["Stay","Churn"],autopct="%1.1f%%",
                      colors=["lime","crimson"],startangle=90)
            ax[1].set_title("Churn Distribution")
            plt.show()

def show_metrics(change=None):
    with output:
        clear_output(wait=True)
        if mode_dropdown.value=="Regression (MonthlyCharges)":
            print("Regression Metrics")
            print(f"R2 Score: {r2_score(y_test_reg,reg_pred):.2f}")
        else:
            print("Classification Metrics")
            print(classification_report(y_test_cls,cls_pred))

def reset_dashboard(change=None):
    tenure_slider.value=24
    charges_slider.value=75
    senior_toggle.value=0
    mode_dropdown.value="Classification (Churn)"
    update_dashboard()

predict_button.on_click(update_dashboard)
metrics_button.on_click(show_metrics)
reset_button.on_click(reset_dashboard)

# --- Layout ---
dashboard = widgets.VBox([
    widgets.HBox([tenure_slider,charges_slider,senior_toggle]),
    mode_dropdown,
    widgets.HBox([predict_button,metrics_button,reset_button]),
    output
])
display(dashboard)

# Initial run
update_dashboard()
```
