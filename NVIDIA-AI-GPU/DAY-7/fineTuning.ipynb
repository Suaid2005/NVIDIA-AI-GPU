```python
import torch
import time
import psutil

from transformers import (
    pipeline,
    AutoTokenizer,
    AutoModel,
    AutoModelForCausalLM
)

print("=" * 60)
print("HUGGING FACE AI WORKSTATION")
print("=" * 60)

# ===================================================
# Device Information
# ===================================================

device = "cuda" if torch.cuda.is_available() else "cpu"

print("\nDEVICE INFORMATION")
print("Device:", device)

if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))

# ===================================================
# Tokenization Demo
# ===================================================

print("\nTOKENIZATION")

bert_tokenizer = AutoTokenizer.from_pretrained(
    "bert-base-uncased"
)

text = """
Artificial Intelligence is transforming
education and cybersecurity.
"""

tokens = bert_tokenizer.tokenize(text)

print("Tokens:")
print(tokens)

print("Token Count:", len(tokens))

# ===================================================
# Embedding Demo
# ===================================================

print("\nEMBEDDINGS")

bert_model = AutoModel.from_pretrained(
    "bert-base-uncased"
).to(device)

inputs = bert_tokenizer(
    text,
    return_tensors="pt",
    padding=True,
    truncation=True
)

inputs = {
    k: v.to(device)
    for k, v in inputs.items()
}

with torch.no_grad():
    outputs = bert_model(**inputs)

embedding = outputs.last_hidden_state

print(
    "Embedding Shape:",
    embedding.shape
)

# Sentence embedding
sentence_embedding = embedding.mean(dim=1)

print(
    "Sentence Embedding Shape:",
    sentence_embedding.shape
)

# ===================================================
# Text Generation
# ===================================================

print("\nTEXT GENERATION")

generator = pipeline(
    "text-generation",
    model="distilgpt2",
    device=0 if torch.cuda.is_available() else -1
)

result = generator(
    "Machine Learning is",
    max_new_tokens=50,
    do_sample=True,
    temperature=0.8
)

print(
    result[0]["generated_text"]
)

# ===================================================
# Sentiment Analysis
# ===================================================

print("\nSENTIMENT ANALYSIS")

sentiment = pipeline(
    "sentiment-analysis",
    device=0 if torch.cuda.is_available() else -1
)

reviews = [
    "This workshop is amazing.",
    "The session was boring."
]

for review in reviews:
    print("\nReview:", review)
    print(sentiment(review))

# ===================================================
# Summarization
# ===================================================

print("\nSUMMARIZATION")

summarizer = pipeline(
    "summarization",
    model="facebook/bart-large-cnn",
    device=0 if torch.cuda.is_available() else -1
)

article = """
Artificial Intelligence is a branch of computer science focused on creating
systems capable of performing tasks that typically require human intelligence.
These tasks include learning, reasoning, problem-solving, perception,
language understanding, and decision-making. AI technologies are being used
across industries such as healthcare, finance, cybersecurity, education,
and autonomous systems.
"""

summary = summarizer(
    article,
    max_length=40,
    min_length=10,
    do_sample=False
)

print(
    summary[0]["summary_text"]
)

# ===================================================
# Translation
# ===================================================

print("\nTRANSLATION")

translator = pipeline(
    "translation_en_to_fr",
    model="t5-small",
    device=0 if torch.cuda.is_available() else -1
)

translated = translator(
    "Artificial Intelligence is transforming education."
)

print(
    translated[0]["translation_text"]
)

# ===================================================
# Throughput Benchmark
# ===================================================

print("\nBENCHMARK")

prompt = "Explain Transformers in AI."

start = time.time()

output = generator(
    prompt,
    max_new_tokens=100
)

end = time.time()

generated_text = output[0]["generated_text"]

token_count = len(
    generator.tokenizer.encode(
        generated_text
    )
)

throughput = token_count / (end - start)

print(
    "Execution Time:",
    round(end - start, 2),
    "seconds"
)

print(
    "Tokens Generated:",
    token_count
)

print(
    "Tokens/sec:",
    round(throughput, 2)
)

# ===================================================
# Memory Monitoring
# ===================================================

print("\nMEMORY USAGE")

process = psutil.Process()

ram_usage = (
    process.memory_info().rss
    / 1024**2
)

print(
    "RAM Usage:",
    round(ram_usage, 2),
    "MB"
)

# ===================================================
# GPU Monitoring
# ===================================================

if torch.cuda.is_available():

    gpu_memory = (
        torch.cuda.memory_allocated()
        / 1024**3
    )

    print(
        "GPU Memory Used:",
        round(gpu_memory, 2),
        "GB"
    )

# ===================================================
# Completion
# ===================================================

print("\nExecution Complete")
```

```python
# =====================================================
# LoRA Fine-Tuning Demo with DistilGPT2
# =====================================================

import os
os.environ["WANDB_DISABLED"] = "true"

import torch

from datasets import Dataset

from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    Trainer,
    TrainingArguments,
    DataCollatorForLanguageModeling
)

from peft import (
    LoraConfig,
    get_peft_model,
    PeftModel
)

# =====================================================
# 1. Load Model & Tokenizer
# =====================================================

model_name = "distilgpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)

tokenizer.pad_token = tokenizer.eos_token

base_model = AutoModelForCausalLM.from_pretrained(
    model_name
)

# =====================================================
# 2. Configure LoRA
# =====================================================

lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    lora_dropout=0.1,
    bias="none",
    task_type="CAUSAL_LM"
)

model = get_peft_model(
    base_model,
    lora_config
)

print("\nTrainable Parameters")
model.print_trainable_parameters()

# =====================================================
# 3. Create Dataset
# =====================================================

texts = [
    "AI is transforming industries.",
    "Machine learning helps computers learn from data.",
    "LoRA enables efficient fine tuning of language models.",
    "Large language models generate human like text."
]

dataset = Dataset.from_dict(
    {"text": texts}
)

# =====================================================
# 4. Tokenization
# =====================================================

def tokenize_function(examples):

    result = tokenizer(
        examples["text"],
        truncation=True,
        padding="max_length",
        max_length=64
    )

    result["labels"] = result["input_ids"].copy()

    return result


tokenized_dataset = dataset.map(
    tokenize_function,
    batched=True
)

# =====================================================
# 5. Training Arguments
# =====================================================

training_args = TrainingArguments(
    output_dir="./lora_output",
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=2,
    logging_steps=1,
    save_strategy="no",
    report_to="none"
)

# =====================================================
# 6. Trainer
# =====================================================

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=DataCollatorForLanguageModeling(
        tokenizer=tokenizer,
        mlm=False
    )
)

# =====================================================
# 7. Train
# =====================================================

print("\nStarting LoRA Training...\n")

trainer.train()

# =====================================================
# 8. Save Adapter
# =====================================================

adapter_path = "./lora_adapter"

model.save_pretrained(
    adapter_path
)

print("\nLoRA adapter saved!")

# =====================================================
# 9. Load Adapter Back
# =====================================================

base_model = AutoModelForCausalLM.from_pretrained(
    model_name
)

finetuned_model = PeftModel.from_pretrained(
    base_model,
    adapter_path
)

print("\nAdapter Loaded Successfully!")

# =====================================================
# 10. Test Generation
# =====================================================

prompt = "AI is"

inputs = tokenizer(
    prompt,
    return_tensors="pt"
)

with torch.no_grad():

    outputs = finetuned_model.generate(
        **inputs,
        max_new_tokens=30,
        do_sample=True,
        temperature=0.8
    )

generated_text = tokenizer.decode(
    outputs[0],
    skip_special_tokens=True
)

print("\nGenerated Text:")
print(generated_text)

print("\nLoRA Fine-Tuning Completed!")
```

```python
# ==========================================================
# LoRA BENCHMARKING SCRIPT
# ==========================================================

import os
import time
import torch
import pandas as pd

os.environ["WANDB_DISABLED"] = "true"

from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    TrainingArguments,
    Trainer,
    DataCollatorForLanguageModeling
)

from peft import (
    LoraConfig,
    get_peft_model
)

# ==========================================================
# DEVICE
# ==========================================================

device = "cuda" if torch.cuda.is_available() else "cpu"

print("=" * 60)
print("LoRA Benchmark")
print("=" * 60)

print("Device:", device)

if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))

# ==========================================================
# DATASET
# ==========================================================

texts = [
    "AI is transforming industries.",
    "Machine learning helps computers learn from data.",
    "LoRA enables efficient fine tuning of language models.",
    "Large language models generate human like text."
] * 250

dataset = Dataset.from_dict(
    {"text": texts}
)

# ==========================================================
# TOKENIZER
# ==========================================================

model_name = "distilgpt2"

tokenizer = AutoTokenizer.from_pretrained(
    model_name
)

tokenizer.pad_token = tokenizer.eos_token

# ==========================================================
# TOKENIZATION
# ==========================================================

def tokenize_function(examples):

    tokens = tokenizer(
        examples["text"],
        truncation=True,
        padding="max_length",
        max_length=64
    )

    tokens["labels"] = tokens["input_ids"].copy()

    return tokens


tokenized_dataset = dataset.map(
    tokenize_function,
    batched=True
)

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
)

# ==========================================================
# CONFIGURATIONS
# ==========================================================

configs = [
    {"r": 4, "alpha": 8},
    {"r": 8, "alpha": 16},
    {"r": 16, "alpha": 32},
    {"r": 32, "alpha": 64},
]

results = []

# ==========================================================
# BENCHMARK LOOP
# ==========================================================

for config in configs:

    r = config["r"]
    alpha = config["alpha"]

    print("\n" + "=" * 60)
    print(f"Testing Rank={r} Alpha={alpha}")
    print("=" * 60)

    # ------------------------------------------------------
    # Clear GPU Memory
    # ------------------------------------------------------

    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.cuda.reset_peak_memory_stats()

    # ------------------------------------------------------
    # Load Model
    # ------------------------------------------------------

    base_model = AutoModelForCausalLM.from_pretrained(
        model_name
    )

    # ------------------------------------------------------
    # LoRA Config
    # ------------------------------------------------------

    lora_config = LoraConfig(
        r=r,
        lora_alpha=alpha,
        lora_dropout=0.1,
        bias="none",
        task_type="CAUSAL_LM"
    )

    model = get_peft_model(
        base_model,
        lora_config
    )

    model.to(device)

    # ------------------------------------------------------
    # Parameter Count
    # ------------------------------------------------------

    trainable_params = sum(
        p.numel()
        for p in model.parameters()
        if p.requires_grad
    )

    total_params = sum(
        p.numel()
        for p in model.parameters()
    )

    print(f"Trainable Params: {trainable_params:,}")
    print(f"Total Params: {total_params:,}")

    # ------------------------------------------------------
    # Training Arguments
    # ------------------------------------------------------

    training_args = TrainingArguments(
        output_dir=f"./output_r{r}",
        overwrite_output_dir=True,
        num_train_epochs=2,
        per_device_train_batch_size=8,
        save_strategy="no",
        logging_steps=50,
        report_to="none"
    )

    # ------------------------------------------------------
    # Trainer
    # ------------------------------------------------------

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=tokenized_dataset,
        data_collator=data_collator
    )

    # ------------------------------------------------------
    # Train
    # ------------------------------------------------------

    start_time = time.time()

    train_output = trainer.train()

    training_time = time.time() - start_time

    final_loss = train_output.training_loss

    # ------------------------------------------------------
    # Peak GPU Memory
    # ------------------------------------------------------

    if torch.cuda.is_available():

        peak_gpu_memory = (
            torch.cuda.max_memory_allocated()
            / 1024**2
        )

    else:

        peak_gpu_memory = 0

    # ------------------------------------------------------
    # Generation Test
    # ------------------------------------------------------

    prompt = "AI is"

    inputs = tokenizer(
        prompt,
        return_tensors="pt"
    )

    inputs = {
        k: v.to(device)
        for k, v in inputs.items()
    }

    model.eval()

    with torch.no_grad():

        output = model.generate(
            **inputs,
            max_new_tokens=20,
            do_sample=True,
            temperature=0.8,
            pad_token_id=tokenizer.eos_token_id
        )

    generated_text = tokenizer.decode(
        output[0],
        skip_special_tokens=True
    )

    # ------------------------------------------------------
    # Print Results
    # ------------------------------------------------------

    print("\nGenerated Text:")
    print(generated_text)

    print("\nTraining Time:",
          round(training_time, 2),
          "seconds")

    print("Peak GPU Memory:",
          round(peak_gpu_memory, 2),
          "MB")

    print("Final Loss:",
          round(final_loss, 4))

    # ------------------------------------------------------
    # Save Benchmark Result
    # ------------------------------------------------------

    results.append({
        "Rank": r,
        "Alpha": alpha,
        "Trainable_Params": trainable_params,
        "Training_Time_sec": round(training_time, 2),
        "Peak_GPU_MB": round(peak_gpu_memory, 2),
        "Final_Loss": round(final_loss, 4)
    })

# ==========================================================
# RESULTS TABLE
# ==========================================================

print("\n")
print("=" * 80)
print("FINAL BENCHMARK RESULTS")
print("=" * 80)

df = pd.DataFrame(results)

print(df)

# ==========================================================
# BEST CONFIG
# ==========================================================

best = df.sort_values(
    by="Final_Loss"
).iloc[0]

print("\nBest Configuration:")
print(best)

print("\nBenchmark Completed Successfully!")
```
